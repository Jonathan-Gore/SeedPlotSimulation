## Example of Rejection Sampling ##

```python
"""
Let's say we're at x=0 (left side)
~$ prob = np.exp(-1.0 * 0 / width) = 1.0

Almost all points will be kept since random < 1.0 is true ~100% of time

Now at x = width (right side)
~$ prob = np.exp(-1.0 * width / width) = np.exp(-1) â‰ˆ 0.37

Only ~37% of points will be kept since random < 0.37 is true ~37% of time
This creates the gradient because:

* Points with high probability (left side) are kept more often
* Points with low probability (right side) are rejected more often
* Rejections are ultimately made by a random number generator, but influenced by a set of probabilistic rules
"""
```

```python
import numpy as np
import matplotlib.pyplot as plt

def visualize_rejection_sampling():
    width = 100
    points_accepted = []
    points_rejected = []
    
    # Try 1000 random points
    for _ in range(1000):
        x = np.random.uniform(0, width)
        y = np.random.uniform(0, width)
        
        # Calculate probability (using x-direction gradient)
        prob = np.exp(-1.0 * x / width)
        
        # Rejection sampling
        if np.random.random() < prob:
            points_accepted.append((x, y))
        else:
            points_rejected.append((x, y))
    
    # Convert to arrays for plotting
    accepted = np.array(points_accepted)
    rejected = np.array(points_rejected)
    
    plt.figure(figsize=(10, 5))
    
    # Plot rejected points in red, accepted in blue
    if len(rejected) > 0:
        plt.scatter(rejected[:, 0], rejected[:, 1], c='red', alpha=0.2, label='Rejected')
    if len(accepted) > 0:
        plt.scatter(accepted[:, 0], accepted[:, 1], c='blue', alpha=0.5, label='Accepted')
    
    plt.title('Rejection Sampling Visualization')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.show()

visualize_rejection_sampling()
```
